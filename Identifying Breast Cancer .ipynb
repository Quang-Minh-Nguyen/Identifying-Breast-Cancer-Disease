## 4. Check Null Values 

```python
data.isnull().sum()
```

#### As we can see, there are irrelavant columns which are 'id' and 'Unnamed: 32'. Therefore, I will drop them.

```python
data = data.drop(['Unnamed: 32'], axis =1)
data = data.drop(['id'], axis =1)
```

## 5. Check Duplicated Data 

```python
data.duplicated().sum()
```

#### There is no duplicated data. I do not have to drop any.

## 6. Convert Categotical Value

```python
cate_val = []
cont_val = []

for column in data.columns:
    if data[column].nunique() <= 10:
        cate_val.append(column)
    else:
        cont_val.append(column)
```

```python
cate_val
```

```python
cont_val
```

```python
column_to_move = data.pop('diagnosis')
data.insert(30, 'diagnosis', column_to_move)
```

```python
data
```

```python
data['diagnosis'] = data['diagnosis'].map({'M': 1, 'B':0})
```

```python
data['diagnosis'].value_counts()
```

```python
sns.countplot(data['diagnosis'])
```

#### As we can see, the number of people who do not have the breast cancer is about 150% larger than the number of people who do have the disease. Therefore, I will handle the imbalance by using overfitting  

## 7. Handle Imbalanced Data With SMOTE

```python
from imblearn.over_sampling import SMOTE
```

```python
X = data.drop('diagnosis', axis =1)
y = data['diagnosis']
```

```python
X_res, y_res = SMOTE().fit_resample(X,y)
```

```python
y_res.value_counts()
```

## 8.Overall Statistics About the Dataset

```python
data.describe()
```

## 9. Draw Correlation Matrix

```python
plt.figure(figsize=(21,10))
sns.heatmap(data.corr(), annot=True)
```

## 10. Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
```

```python
st = StandardScaler()
data[cont_val] = st.fit_transform(data[cont_val])
```

```python
data.head()
```

## 11. Spliting The Dataset Into The Training Set And Test Set

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_res,y_res, test_size =0.20,random_state = 42)
```

## 12. Logistics Regression

```python
data.head()
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr = LogisticRegression()
lr.fit(X_train, y_train)
```

```python
y_pred_lr = lr.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test, y_pred_lr)
```

## 13. SVC

```python
from sklearn import svm
```

```python
svm = svm.SVC()
```

```python
svm.fit(X_train, y_train)
```

```python
y_pred_svm = svm.predict(X_test)
```

```python
accuracy_score(y_test, y_pred_svm)
```

## 14. KNeighbors Classifier

```python
from sklearn.neighbors import KNeighborsClassifier 
```

```python
kn = KNeighborsClassifier() 
```

```python
kn.fit(X_train, y_train)
```

```python
y_pred_kn = kn.predict(X_test)
```

```python
accuracy_score(y_test, y_pred_kn)
```

```python
score = []

for k in range(1,40):
    kn = KNeighborsClassifier(n_neighbors = k)
    kn.fit(X_train, y_train)
    y_pred_kn = kn.predict(X_test)
    score.append(accuracy_score(y_test, y_pred_kn))
```

```python
score
```

#### k = 3 is still the best option  

```python
kn = KNeighborsClassifier(n_neighbors = 3)
kn.fit(X_train, y_train)
y_pred_kn = kn.predict(X_test)
accuracy_score(y_test, y_pred_kn)
```

## 15. Non-Linear ML Algorithms

#### For non-linear ML algorithms, preprocessing is not necessary.

```python
data1 = pd.read_csv('raw_data.csv')
```

```python
data1
```

```python
data1 = data1.drop(['Unnamed: 32'], axis =1)
data1 = data1.drop(['id'], axis =1)
```

## 19. Model Comparation

```python
final_data = pd.DataFrame({'Model':['LR','SVM','KN','DT','RF','GB'],
                          'ACC':[accuracy_score(y_test, y_pred_lr),
                                accuracy_score(y_test, y_pred_svm),
                                accuracy_score(y_test, y_pred_kn),
                                accuracy_score(y1_test, y_pred_dt),
                                accuracy_score(y1_test, y_pred_rf),
                                accuracy_score(y1_test, y_pred_gb)]})
```
